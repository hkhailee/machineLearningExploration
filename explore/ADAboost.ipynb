```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('/content/gdrive/MyDrive/Hailee Kiesecker/Final-Due5 3@12/heart_failure_clinical_records_dataset.csv')
```

```python
y = data['DEATH_EVENT']
```

```python
X = data.drop(columns={'DEATH_EVENT'})
X=StandardScaler().fit_transform(X)
```

```python
cv = KFold(n_splits=20, random_state=33, shuffle=True)
model = LogisticRegression()

scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
model02 = LogisticRegression()

clf = LogisticRegression(random_state=0, max_iter=5000).fit(X, y)
model02.fit(X, y)

y_preds = clf.predict(X[:, :])
target_names = ['class 0 ', 'class 1']
print(classification_report(y, y_preds, target_names=target_names))
```

Using K-fold cross validation with logistic regression to stimulate a AdaBoost for our "weak" learning model, we were able to get better trained classification logistic models than a standard one time logistic regression in terms of accuracy and precision
